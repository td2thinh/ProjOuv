```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
plt.rcParams["figure.autolayout"] = True
```

```python
df = pd.read_csv('./random.csv')
```

```python
df_time_nbits = df[['time1', 'nbits', 'time2']]
```

```python
plt.plot(df_time_nbits['nbits'], df_time_nbits['time1'])
plt.plot(df_time_nbits['nbits'], df_time_nbits['time2'])
plt.legend(['CompressionList', 'CompressionTree'])
plt.xlabel('nombre de bits')
plt.ylabel('time (s)')
```

```python
plt.plot(df_time_nbits['nbits'], df_time_nbits['time1'])
plt.legend(['CompressionList'])
plt.xlabel('nombre de bits')
plt.ylabel('time (s)')
```

```python
plt.plot(df_time_nbits['nbits'], df_time_nbits['time2'], color='orange')
plt.legend(['CompressionTree'])
plt.xlabel('nombre de bits')
plt.ylabel('time (s)')
```
